#### Task 1: Data Quality Report Generator 

```python
# Task: 
# Write a function that takes a list of values and returns the count of how many values are None or empty strings.

# Solution:
def count_missing_values(data_list):
    return data_list.count(None) + data_list.count("")

total_missing_values = count_missing_values([25, 30, None, 45, "", 50, 35, None, 40, 28, 32, "", 38, 42, 29])
print("Total missing values:", total_missing_values)

#  Calculate what percentage of the data is NOT missing.

# Solution:
def calculate_completeness_percentage(data_list):
    missing_values = count_missing_values(data_list)
    return (len(data_list) - missing_values) / len(data_list) * 100

data_completeness_percentage = calculate_completeness_percentage([25, 30, None, 45, "", 50, 35, None, 40, 28, 32, "", 38, 42, 29])
print("data_completeness_percentage:", data_completeness_percentage)

# Create a full report about the data quality.

# Solution:
customer_ages = [25, 30, None, 45, "", 50, 35, None, 40, 28, 32, "", 38, 42, 29]

def generate_data_quality_report(data_list, column_name):
    total_records = len(data_list)
    missing_values = count_missing_values(data_list)
    complete_values = total_records - missing_values
    completeness_percentage = calculate_completeness_percentage(data_list)


    print("=== Data Quality Report ===")
    print(f"Column: {column_name}")
    print(f"Total records: {total_records}")
    print(f"Missing values: {missing_values}")
    print(f"Complete values: {complete_values}")
    print(f"Completeness percentage: {completeness_percentage:.2f}%")
    
    
data_quality_report = generate_data_quality_report(customer_ages, "Customer Ages")
```
